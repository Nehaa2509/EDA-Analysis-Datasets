```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("/content/sample_data/iris_dataset.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
df.hist(figsize=(10,6))
plt.show()

```

```python
sns.countplot(x="species", data=df)
plt.show()
```

```python
sns.boxplot(data=df)
plt.show()
```

```python
numeric_df = df.select_dtypes(include=["int64", "float64"])
numeric_df


```

```python
plt.figure(figsize=(9,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.show()
```

### EDA Summary
- Dataset explored using statistical and visual methods
- Outliers identified using box plots
- Correlation analysis performed
- Important features identified for prediction


```python
df = pd.read_csv("/content/sample_data/netflix_movies_tv_shows.csv")
df.head()
```

```python
df.info()
df.describe(include="all")
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["release_year"], bins=20)
plt.show()
```

```python
sns.countplot(x="type", data=df)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(y="rating", data=df)
plt.show()
```

```python
sns.boxplot(x="type", y="release_year", data=df)
plt.show()
```

```python
numeric_df = df.select_dtypes(include="number")

plt.figure(figsize=(6,4))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.show()
```

### EDA Summary  Netflix Dataset
- Majority of Netflix content consists of movies
- Content additions increased significantly after 2015
- TV-MA and TV-14 are dominant ratings
- Dataset is categorical-heavy with limited numerical correlation
- Features like type, rating, and listed_in are important for prediction
